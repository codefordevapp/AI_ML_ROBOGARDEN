```python
from typing import List
from datetime import datetime

class BankAccount(object):
    
    def __init__(self, name:str, accountNumber:int, accountType:str, balance:float = 0)->None:
        self.name = name
        self.accountNumber = accountNumber
        self.accountType = accountType
        self.balance = balance
        self.filename = str(self.accountNumber) + "_" + self.accountType + "_" + self.name + ".txt"
  
    def deposit(self, money:float)->None:
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        self.balance += money
        with open(self.filename, 'a') as file:
            file.write(f'{date}, deposit :  {money}\n')

    def withdraw(self, money):
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        if self.balance > money:
            self.balance -= money
            with open(self.filename, 'a') as file:
                file.write(f'{date}, withdraw : {money}\n')
        else:
            print(f'the amount {money} you need is not available')
            with open(self.filename, 'a') as file:
                file.write(f'{date}, withdraw {money} : failed\n')

    def get_balance(self):
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        with open(self.filename, 'a') as file:
            file.write(f'{date}, checking balance : {self.balance}\n')
        print(self.balance)
        return self.balance

    def get_account_type(self):
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        with open(self.filename, 'a') as file:
            file.write(f'{date}, checking account type : "{self.accountType}"\n')
        print(self.accountType)
        return self.accountType

    def get_account_number(self):
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        with open(self.filename, 'a') as file:
            file.write(f'{date}, checking account number :"{self.accountNumber}"\n')
        print(self.accountNumber)
        return self.accountNumber

    def get_holder_name(self):
        date = datetime.today().strftime('%Y-%m-%d %H:%M:%S')
        with open(self.filename, 'a') as file:
            file.write(f'{date}, checking account name : "{self.name}"\n')
        print(self.name)
        return self.name

    def transactions_story(self):
        with open(self.filename, 'r') as file:
            print(file.read())

#create account
Tom = BankAccount('Tom', 1000, 'checking')
Marcia = BankAccount('Marcia', 1001, 'checking')
Delia = BankAccount('Delia', 1002, 'saving')
Bob = BankAccount('Bob', 1003, 'checking')
John = BankAccount('John', 1004, 'saving')
Victor = BankAccount('Victor', 1005, 'saving')
David = BankAccount('David', 1006, 'checking')

#deposit money
Tom.deposit(852)
'''
Marcia.deposit(1000)
Delia.deposit(500)
Bob.deposit(600)
John.deposit(350)
Victor.deposit(275)
David.deposit(655)
'''

#check balance
Tom.get_balance()
'''
Marcia.get_balance()
Delia.get_balance()
Bob.get_balance()
John.get_balance()
Victor.get_balance()
David.get_balance()
'''

#withdraw money
Tom.withdraw(300)
'''
Marcia.withdraw(1200)
Delia.withdraw(200)
Bob.withdraw(100)
John.withdraw(50)
Victor.withdraw(275)
David.withdraw(700)
'''

#check account type
Tom.get_account_type()
'''
Marcia.get_account_type()
Delia.get_account_type()
Bob.get_account_type()
John.get_account_type()
Victor.get_account_type()
David.get_account_type()
'''

#retrieve account number
Tom.get_account_number()
'''
Marcia.get_account_number()
Delia.get_account_number()
Bob.get_account_number()
John.get_account_number()
Victor.get_account_number()
David.get_account_number()
'''

#retrieve holder name
Tom.get_holder_name()
'''
Marcia.get_holder_name()
Delia.get_holder_name()
Bob.get_holder_name()
John.get_holder_name()
Victor.get_holder_name()
David.get_holder_name()
'''

#check transaction
Tom.transactions_story()
'''
Marcia.transactions_story()
Delia.transactions_story()
Bob.transactions_story()
John.transactions_story()
Victor.transactions_story()
David.transactions_story()
'''

```
