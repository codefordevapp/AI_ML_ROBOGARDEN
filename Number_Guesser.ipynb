```python
from typing import List
import random

class Chatbot:
    
    def __init__(self, number_of_guess:int  = 4, number_of_hints:int = 3, min_guess:int = 0, max_guess:int = 100):
        self.number_of_guess = number_of_guess
        self.number_of_hints = number_of_hints
        self.min_guess = min_guess
        self.max_guess = max_guess
        self.__number = random.randint(self.min_guess, self.max_guess)
        self.__factors = self.__all_factors()
        self.__multiples = self.__all_multiples()
        self.__larger = self.__all_larger()
        self.__smaller = self.__all_smaller()
        self.__parity = self.__num_parity()
    
    def greeting(self)->None:
        print(f'''
Welcome to number guess game!
I have guessed a number between {self.min_guess} and {self.max_guess} and you have 4 guesses to discover my number.
You can request 3 hints by just typing a character as explain below:
    'a' : HINT to get a factor or a multiple of the number
    'b' : HINT to get a larger or a smaller of the number
    'c' : HINT to get the parity of the number
             ''')
    
    def user_guess_num(self)->int:
        while True:
            try:
                user_guess = int(input('\nPlease enter your guess:'))
                if (user_guess > self.max_guess or user_guess < self.min_guess):
                    print(f'\nMy number is between {self.min_guess} and {self.max_guess}')
                else:
                    return user_guess
            except:
                print(f'\nEnter an integer value between {self.min_guess} and {self.max_guess}')
            
    
    def correct_guess(self)->None: 
        congrats = ['\nCONGRATULATION, YOUR GUESS IS RIGHT!', '\nCONGRATULATION, YOUR ARE AMAZING!','\nFANTASTIC, VERY GOOD GUESS!']
        i = random.randint(0, 2)
        print(congrats[i])
        
    def wrong_guess(self):
        self.number_of_guess -= 1
        if self.number_of_guess > 0:
            uplift = ["\nYOU'RE CLOSE!", "\nALMOST!","\nNOT SO BAD!"]
            i = random.randint(0, 2)
            print(f"\n{uplift[i]} Guess one more time, you still have {self.number_of_guess} guess(es)")
            if self.number_of_hints > 0:
                while self.number_of_hints > 0:
                    print(f'\nyou can request HINT, you still have {self.number_of_hints} hint(s)')
                    try:
                        need_hint = int(input('\nDo you need a HINT ? type 1 for yes, 0 if not : '))
                        if need_hint:
                            user_hint = input("\nWhat type of HINT do you need: a, b or c? ")
                            if user_hint in ['a', 'b', 'c']:
                                if user_hint == 'a':
                                    self.hint_a()
                                elif user_hint == 'b':
                                    self.hint_b()
                                else:
                                    self.hint_c()
                            else:
                                print('\nEnter the correct HINT')
                        else:
                            break 
                    except:
                        print('\nEnter the correct choice : 1 = need HINT, 0  = no need')
            else:
                print(f'\nsorry, you run out your {self.number_of_hints} hints')                            
        else:
            print(f'\nYou run out your number of guesses, the number is {self.__number}')


    def __all_factors(self)->List:
        if self.__number > 0:
            self.__factors = [1, self.__number]
            i = 2
            while i*i < self.__number:
                if self.__number % i == 0:
                    self.__factors.extend([i, self.__number // i])
                i = i + 1
            if i*i == self.__number:
                self.__factors.append(i)
        else:
            self.__factors = []
        return self.__factors

    def __all_multiples(self)->List:
        if self.__number > 0:
            #all multiples of chatboat guess number
            self.__multiples = []
            i = 2
            while i * self.__number <  self.max_guess:
                self.__multiples.append(i * self.__number)
                i = i + 1
        else:
            self.__multiples = []
        return self.__multiples
            
    def hint_a(self)->None:
        self.number_of_hints -= 1
        if self.__factors == [] and self.__multiples == []:
            print("\nthe number don't have any factors and any multiples")
        elif self.__factors == []:
            m = len(self.__multiples)
            i = random.randint(0, m - 1)
            print(f"\n{self.__multiples[i]} is a multiple of the number you guess")
        elif self.__multiples == []:
            f = len(self.__factors)
            j = random.randint(0, f - 1)
            print(f"\n{self.__factors[j]} is a factor of the number you guess")
        else:
            m = len(self.__multiples)
            f = len(self.__factors)
            choice = random.randint(0, 1)
            if choice == 0:
                m = len(self.__multiples)
                i = random.randint(0, m - 1)
                print(f"\n{self.__multiples[i]} is multiple of the number you guess")
            else:
                f = len(self.__factors)
                j = random.randint(0, f - 1)
                print(f"\n{self.__factors[j]} is a factor of the number you guess")

    def __all_smaller(self)->List:
        #all numbers smaller than the chatboat guess number
        smaller = [i for i in range(self.min_guess, self.__number)]
        return smaller
        
    def __all_larger(self)->List:
        #all numbers larger than the chatboat guess number
        larger = [i for  i in range(self.__number + 1, self.max_guess + 1)]
        return larger
   
    def hint_b(self)->None:
        self.number_of_hints -= 1
        if self.__smaller == []:
            l = len(self.__larger)
            i = random.randint(0, l - 1)
            print(f"\n{self.__larger[i]} is larger than the number you guess")
        elif self.__larger == []:
            s = len(self.__smaller)
            j = random.randint(0, s - 1)
            print(f"\n{self.__smaller[j]} is smaller than the number you guess")
        else:
            s = len(self.__smaller)
            l = len(self.__larger)
            choice = random.randint(0, 1)
            if choice == 0:
                s = len(self.__smaller)
                i = random.randint(0, s - 1)
                print(f"\n{self.__smaller[i]} is smaller than the number you guess")
            else:
                l = len(self.__larger)
                j = random.randint(0, l - 1)
                print(f"\n{self.__larger[j]} is larger than the number you guess")

    def __num_parity(self)->bool:
        return self.__number % 2 == 0

    def hint_c(self)->None:
        self.number_of_hints -= 1
        if self.__num_parity():
            print('\nthe number you guess is even')
        else:
            print('\nthe number you guess is odd')
        
    def game(self)->None:
        self.greeting()
        while self.number_of_guess > 0:
            user_guess = self.user_guess_num()
            if self.__number == user_guess:
                self.correct_guess()
                break
            else:
                self.wrong_guess()


play = Chatbot()
play.game()
while True:
    play_again = int(input('\nWould you like a new game? type "1" for yes, "0" otherwise'))
    if play_again in [0, 1]:
        if play_again:
            play = Chatbot()
            play.game()
        else:
            print('\nYOU ARE A GOOD GUESSER, SEE YOU SOON!')
            break
    else:
        print('\nPlease enter the correct choice')



    
        
```
